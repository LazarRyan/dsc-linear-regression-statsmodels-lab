## Step 2: Run a Simple Linear Regression with `TV` as the Predictor

As the analysis above indicates, `TV` looks like it has the strongest relationship with `sales`. Let's attempt to quantify that using linear regression.

```python
# Import libraries
from sklearn.linear_model import LinearRegression
# Determine X and y values
X = df[['TV']]  # Predictor variable (TV)
y = df['sales']  # Target variable
# Create an OLS model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)
```

```python
# Make predictions
predictions = model.predict(X)
```

```python
# Plot the data points
plt.scatter(X, y, label='Data')

# Plot the regression line
plt.plot(X, predictions, color='red', label='Regression Line')

# Label the axes and add a legend
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.legend()

# Show the plot
plt.show()
```

## Step 3: Evaluate and Interpret Results from Step 2

## Step 4: Visualize Model with `TV` as Predictor

Create at least one visualization that shows the prediction line against a scatter plot of `TV` vs. sales, as well as at least one visualization that shows the residuals.

```python
# Plot the model fit (scatter plot and regression line)
# Make predictions
predictions = model.predict(X)

# Visualization 1: Scatter plot of TV vs. Sales with Prediction Line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Actual Data', alpha=0.7)
plt.plot(X, predictions, color='red', linewidth=2, label='Regression Line')
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.title('TV vs. Sales with Prediction Line')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot the model residuals
# Calculate the residuals
residuals = y - predictions

# Visualization 2: Residual Plot
plt.figure(figsize=(10, 6))
sns.residplot(X, residuals, lowess=True, color='blue', scatter_kws={'alpha': 0.7})
plt.xlabel('TV Advertising Budget')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.grid(True)
plt.show()
```

## Step 5: Repeat Steps 2-4 with `radio` as Predictor

Compare and contrast the model performance, coefficient value, etc. The goal is to answer the business question described above.

```python
# Run model
# Split the data into predictor (X) and target (y) variables
X = df[['radio']]  # Predictor variable (Radio)
y = df['sales']     # Target variable

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Get the coefficients and intercept
slope_radio = model.coef_[0]
intercept_radio = model.intercept_

# Make predictions
predictions_radio = model.predict(X)

from sklearn.metrics import r2_score, mean_squared_error

# Calculate R-squared and MSE for the Radio model
r2_radio = r2_score(y, predictions_radio)
mse_radio = mean_squared_error(y, predictions_radio)


# Display results
# Print the metrics for the Radio model
print("Radio Model:")
print(f"R-squared: {r2_radio:.2f}")
print(f"Mean Squared Error: {mse_radio:.2f}")


# Print coefficient values for both models
print("TV Model:")
print(f"Slope (TV Coefficient): {slope_radio:.2f}")
print(f"Intercept: {intercept_radio:.2f}")

print("\nRadio Model:")
print(f"Slope (Radio Coefficient): {slope_radio:.2f}")
print(f"Intercept: {intercept_radio:.2f}")
```

```python
# Visualize model fit
# Split the data into predictor (X) and target (y) variables for TV
X_tv = df[['radio']]  # Predictor variable (TV)
y_tv = df['sales']  # Target variable

# Create a linear regression model for TV
model_tv = LinearRegression()
model_tv.fit(X_tv, y_tv)

# Make predictions for TV
predictions_tv = model_tv.predict(X_tv)

# Visualization: Scatter plot of TV vs. Sales with Prediction Line
plt.figure(figsize=(10, 6))
plt.scatter(X_tv, y_tv, label='Actual Data', alpha=0.7)
plt.plot(X_tv, predictions_tv, color='red', linewidth=2, label='Regression Line')
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.title('Radio vs. Sales with Prediction Line')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Visualize residuals
# Split the data into predictor (X) and target (y) variables for Radio
X_radio = df[['radio']]  # Predictor variable (Radio)
y_radio = df['sales']    # Target variable

# Create a linear regression model for Radio
model_radio = LinearRegression()
model_radio.fit(X_radio, y_radio)

# Make predictions for Radio
predictions_radio = model_radio.predict(X_radio)

# Calculate residuals
residuals_radio = y_radio - predictions_radio

# Visualization: Residual Plot for Radio Model
plt.figure(figsize=(10, 6))
plt.scatter(X_radio, residuals_radio, color='blue', alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--', linewidth=2, label='Zero Residuals Line')
plt.xlabel('Radio Advertising Budget')
plt.ylabel('Residuals')
plt.title('Residual Plot for Radio Model')
plt.legend()
plt.grid(True)
plt.show()
```

On the other hand, the coefficient for radio is much higher. An increase of $1k in radio spending is associated with an increase of sales of about 200 widgets! This is roughly 4x the increase of widget sales that we see for TV.

## Step 6: Repeat Steps 2-4 with `newspaper` as Predictor

Once again, use this information to compare and contrast.

```python
# Run model
# Split the data into predictor (X) and target (y) variables for Newspaper
X_newspaper = df[['newspaper']]  # Predictor variable (Newspaper)
y = df['sales']                   # Target variable

# Create a linear regression model for Newspaper
model_newspaper = LinearRegression()
model_newspaper.fit(X_newspaper, y)

# Make predictions for Newspaper
predictions_newspaper = model_newspaper.predict(X_newspaper)

# Calculate residuals for Newspaper
residuals_newspaper = y - predictions_newspaper

# Display results
# Print the coefficients and intercept for Newspaper
slope_newspaper = model_newspaper.coef_[0]
intercept_newspaper = model_newspaper.intercept_
print("Newspaper Model:")
print(f"Slope (Newspaper Coefficient): {slope_newspaper:.2f}")
print(f"Intercept: {intercept_newspaper:.2f}")
```

```python
# Visualize model fit
# Visualization 1: Scatter plot of Newspaper vs. Sales with Prediction Line
plt.figure(figsize=(10, 6))
plt.scatter(X_newspaper, y, label='Actual Data', alpha=0.7)
plt.plot(X_newspaper, predictions_newspaper, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Newspaper Advertising Budget')
plt.ylabel('Sales')
plt.title('Newspaper vs. Sales with Prediction Line')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Visualize residuals
# Visualization 2: Residual Plot for Newspaper Model
plt.figure(figsize=(10, 6))
plt.scatter(X_newspaper, residuals_newspaper, color='blue', alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--', linewidth=2, label='Zero Residuals Line')
plt.xlabel('Newspaper Advertising Budget')
plt.ylabel('Residuals')
plt.title('Residual Plot for Newspaper Model')
plt.legend()
plt.grid(True)
plt.show()
```

# Your interpretation here
Technically our model and coefficients are still statistically significant at an alpha of 0.05, but the p-values are much higher. For both the F-statistic (overall model significance) and the newspaper coefficient, our p-values are about 0.001, meaning that there is about a 0.1% chance that a variable with no linear relationship would produce these statistics. That is a pretty small false positive rate, so we'll consider the model to be statistically significant and move on to interpreting the other results.

The R-Squared here is the smallest we have seen yet: 0.05. This means that the model explains about 5% of the variance in sales. 5% is well below both the radio model (33%) and the TV model (61%).

The coefficient is also small, though similar to the TV coefficient. An increase of $1k in newspaper spending is associated with about 50 additional widget sales (more precisely, about 54.7). This is still much less than the 200-widget increase associated with $1k of additional radio spending.

Visualizing this model, the best-fit line is clearly not a strong predictor. On the other hand, the residuals exhibit homoscedasticity, meaning that the distribution of the residuals doesn't vary much based on the value of newspaper. This contrasts with the radio and TV residuals which exhibit a "cone" shape, where the errors are larger as the x-axis increases. Homoscedasticity of residuals is a good thing, which we will describe more in depth when we discuss regression assumptions.

Once again, how should we answer the business question? Regardless of the framing, it is unlikely that newspaper is the answer that your stakeholders want. This model has neither the highest R-Squared nor the highest coefficient.
